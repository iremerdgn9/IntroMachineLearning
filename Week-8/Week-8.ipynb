```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
     

iris = load_iris()
iris_data = iris.data
iris_columns = iris.feature_names

iris_df = pd.DataFrame(data=iris_data , columns=iris_columns)
     

print(iris_df.head(30))
```

```python
scaler = MinMaxScaler()
iris_df_scaled = scaler.fit_transform(iris_df)

iris_df_normalized = pd.DataFrame(data=iris_df_scaled, columns=iris_columns)
print(iris_df_normalized.head(30))
```

```python
import seaborn as sns
sns.pairplot(iris_df_normalized)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("C:/Users/User/OneDrive/Masast/KNNAlgorithmDataset.csv")
```

```python
data.head()
```

```python
data.drop(["Unnamed: 32","id"],axis=1, inplace=True)
```

```python
data.tail()
```

```python
M = data[data.diagnosis=="M"]
B = data[data.diagnosis=="B"]
```

```python
plt.scatter(M.radius_mean,M.texture_mean,color="red",label="tehlikeli",alpha=0.7) 
plt.scatter(B.radius_mean,B.texture_mean,color="green",label="zararsz",alpha=0.3)
plt.xlabel("radius_mean(tmor yarap)")
plt.ylabel("texture_mean(tmr dokusu)")
plt.legend()
plt.show()
```

```python
data.diagnosis = [1 if each=="M" else 0 for each in data.diagnosis] 
x_data = data.drop(["diagnosis"],axis=1) 
y = data.diagnosis.values 
```

```python
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
print("accuracy",nb.score(x_test,y_test))  
```
