```python
import seaborn as sns
import matplotlib.pyplot as plt

data= sns.load_dataset("flights")
data = data.pivot("month", "year", "passengers")
 #heat map
sns.heatmap(data = data, annot= True, fmt= "d", cmap="Purples")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")

sns.swarmplot(x="species", y="petal_length", data=data) #swarm plot

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data= sns.load_dataset("titanic")

sns.countplot(data=data, x="class") #bar plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data=sns.load_dataset("tips") #scatter plot

sns.scatterplot(data=data, x="total_bill", y="tip")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data= sns.load_dataset("titanic")

g= sns.FacetGrid(data, col="sex")
g.map(sns.histplot,"age")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris= sns.load_dataset("iris")

g= sns.clustermap(iris.drop("species", axis=1), cmap= "coolwarm", standard_scale=1)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")

# Violin plot
sns.violinplot(x="day", y="total_bill", data=data)
plt.show()
```

```python
import seaborn as sns
sns.set_theme(style="white")

df = sns.load_dataset("penguins")

g = sns.JointGrid(data=df, x="body_mass_g", y="bill_depth_mm", space=0)
g.plot_joint(sns.kdeplot,
             fill=True, clip=((2000, 6500), (15, 20)),
             thresh=0, levels=100, cmap="rocket")
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=30)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

# rnek bir veri seti oluturma
data = pd.DataFrame({
    'age_group': np.repeat(['18-30', '31-45', '46-60', '61+'], 100),
    'feedback_rating': np.random.randint(1, 6, 400)
})

# FacetGrid oluturma
g = sns.FacetGrid(data, col="age_group", margin_titles=True)

# Grafik tipi
g.map(sns.histplot, "feedback_rating")
```
