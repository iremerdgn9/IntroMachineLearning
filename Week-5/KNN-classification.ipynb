```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/User/OneDrive/Masaüstü/KNNAlgorithmDataset.csv")  
data.head()
```

```python
data.info() #kaç satır ve sütun olduğu görünür.
```

```python
radius_mean= data[data.diagnosis=="radius_mean"]  
area_mean= data[data.diagnosis=="area_mean"]           
```

```python
plt.scatter(radius_mean.radius_mean,area_mean.area_mean,color="red",label="radius")
plt.scatter(area_mean.radius_mean,radius_mean.area_mean,color="blue",label="area")
plt.legend() 
plt.xlabel("radius")
plt.ylabel("area")
plt.show()
```

```python
data.diagnosis= [1 if each== "area_mean" else 0 for each in data.diagnosis]
y=data.diagnosis.values
x_data= data.iloc[:,1:6].values
x=(x_data - np.min(x_data))/ (np.max(x_data)-np.min(x_data))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier #KNN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
y_head = knn.predict(x_test)
y_head
```

```python
knn.score(x_test,y_test) 
```
