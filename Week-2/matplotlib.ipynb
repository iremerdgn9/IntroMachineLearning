# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

a= np.arange(0,5,0.4)
b= np.sin(a)
plt.plot(a,b)

plt.title("sin grafiÄŸi")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.random.rand(50)
y= np.random.rand(50)
colors = np.random.rand(50)
sizes= np.random.randint(50,130,50)

plt.scatter(x,y, c=colors, s=sizes)

plt.title("point chart")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data1= np.random.randn(500) #data

plt.hist(data1, bins=70) #graph creat
plt.title("histogram")
plt.xlabel("values")
plt.ylabel("frequency")

plt.show() #show graphic
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= ['a','b','c','d','e','f']
y= np.random.randint(2,5,6)
plt.bar(x,y) #graph creat

plt.title("column chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes= [50,10,40,5,4,9]
plt.pie(sizes)
plt.title("pie chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x=np.arange(-2,4,0.5)
y=np.arange(-2,3,0.5)
x,y = np.meshgrid(x,y)
r= np.sqrt(x**2 + y**2)
z= np.sin(r)

fig= plt.figure()
ax= fig.add_subplot(111, projection= '3d')

ax.plot_surface(x,y,z)
ax.set_title("3D graphics")
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
ax.set_zlabel("z axis")
plt.show()
```
